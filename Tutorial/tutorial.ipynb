# How many Bachelors degrees are awarded in Biological Sciences per year? 
First, create a list of the years where data on Biological Sciences was gathered  
  
Using a `for` loop, we can iterate through the contents of `grad_list` and filter out the entries that are associated with Biological Sciences, using conditional `if` statements.   
  
Then, we can put the `Year` field from that entry into our list

**Fill out the correct keys for accessing the `Major` and the `Year` fields**

```python
bio_years = [] # creating an empty list

for entry in grad_list:
    ## YOUR CODE HERE ##
    if entry['Education']['Major'] == "Biological Sciences": # check the Major of each entry
        year = entry["Year"]
        bio_years.append(year) # add the year of that entry to the list
    ## END ##
print("bio_years = ", bio_years)
```

Expected output: `[1993, 1995, 1997, 1999, 2001, 2003, 2006, 2008, 2010, 2013, 2015]`

It seems that the data we have on Biological Science majors is only for certain years.  
  
Now, lets create a corresponding list of the number of Bachelors degrees awarded for each year

**Fill out the correct keys for accessing the `Bachelors` field**

```python
bio_degs = [] # creating an empty list

for entry in grad_list:
    if entry['Education']['Major'] == "Biological Sciences": # check the Major of each entry
        
        ## YOUR CODE HERE ##
        degs = entry['Education']['Degrees']['Bachelors']
         ## END ##
            
        bio_degs.append(degs) # add the bachelor degrees of that entry to the list
   
print("bio_degs = ",bio_degs)
```

Expected output: `[671374, 708320, 760986, 813139, 108948, 697453, 191812, 146514, 962464, 1099126, 1282386]`
